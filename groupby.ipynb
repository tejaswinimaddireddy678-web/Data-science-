```python
#groupby enables categorywise analysis instead of row by row analysis
```

```python
import pandas as pd
```

```python
df=pd.read_csv("/content/drive/MyDrive/datascience files/weather_by_cities.csv")
```

```python
df
```

```python
#find maximum temperature in each of cities
#find average wind speed per city
```

```python
g=df.groupby("city")
```

```python
for city,city_df in g:
  print(city)
  print(city_df)
```

```python
g.get_group("mumbai")
```

```python
df.columns
```

```python
df.groupby("city")['temperature'].max()
```

```python
#find average wind speed per city
df.groupby("city")["windspeed"].mean()
```

```python
#find the total wind speed for each city
df.groupby("city")["windspeed"].sum()
```

```python
#find the highest wind speed for each city
df.groupby("city")["windspeed"].max()
```
