```python
import numpy as np
```

```python
mat1=np.array([[1,2],[3,4]])
mat2=np.array([[5,6],[7,8]])
```

```python
result=mat1+mat2
print(result)
```

```python
result=mat2-mat1
print(result)
```

```python
result=mat1*mat2
print(result)
```

```python
result=np.matmul(mat1,mat2)
print(result)
result=mat1@mat2
print(result)
result=np.dot(mat1,mat2)
print(result)
```

```python
result=mat1/mat2
print(result)
```

```python
mat=np.array([[1,2,3],[4,5,6]])
print(mat.T)
```

```python
print(mat.shape)
```

```python
identity=np.eye(3)
print(identity)
identity=np.eye(2)
print(identity)
```

```python
mat=np.array([[1,2],[3,4]])
print(np.linalg.det(mat))
```

```python
mat=np.array([[1,2],[3,4]])
print(np.linalg.inv(mat))
```
