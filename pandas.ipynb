```python
import pandas as pd
import numpy as np
```

```python
arr=np.array([10,20,30,40,50])
marks=pd.Series(arr)
print(marks)
```

```python
marks=pd.Series([10,20,30],index=["math","phy","chem"])
print(marks)
print(marks["math"])
print(marks.shape)
print(marks.size)
print(marks.dtype)
print(marks.values)
print(marks.index)
```

```python
numbers=[1,2,34,7]
result=[]
for i in numbers:
  result.append(i*2)
print(result)
```

```python
#vectorized operation means applying one operation to all elements automatically without writing loops
numbers=[1,2,34,7]
arr=np.array(numbers)
print(arr*2)
```

```python
#nan-not a number--- to know the missing values i will use isnull method
data=pd.Series([10,20,30,40,np.nan,50])
print(data.isnull())
print(data.fillna(0))
```

```python
#dataframe
data={"name":["t","e","j","a"],
      "marks":[10,20,30,40],
      "city":["vij","bang","tpt","nlr"]
    }
df=pd.DataFrame(data)
print(df)
print(df.shape)
print(df.size)
print(df.columns)
print(df.dtypes)
print(df.index)
```

```python
df["grade"]=["B","B","B","A"]
print(df)
```

```python
df["marks"]=df["marks"]+20
print(df)
df.drop("city",axis=1)
```
