```python
import seaborn as sns
df=sns.load_dataset('titanic')
df
```

```python
df.head() #its shows 5 values
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df["age"]
```

```python
df.columns
```

```python
df["Age_Mean"]=df["age"].fillna(df["age"].mean())
print(df["Age_Mean"])
```

```python
df[["Age_Mean","age"]]
```

```python
df["Age_Median"]=df["age"].fillna(df["age"].median())
```

```python
df[['Age_Median',"age"]]
```

```python
df["Age_Mode"]=df["age"].fillna(df["age"].mode()[0])
print(df["Age_Mode"])
```

```python
df[["Age_Mode","age"]]
```

```python
df["age"].mode()
```
